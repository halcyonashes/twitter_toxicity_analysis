```python
def analyze_text(client, text):
    analyze_request = {
      'comment': { 'text': text},
      'requestedAttributes': {'TOXICITY': {}, 
                          'SEVERE_TOXICITY': {},
                         'IDENTITY_ATTACK': {}, 
                         'PROFANITY': {}, 
                          'SEXUALLY_EXPLICIT': {}, 
                          'INFLAMMATORY': {}, 
                         'THREAT': {} }
    }
    response = client.comments().analyze(body=analyze_request).execute()
    json_text = json.loads(json.dumps(response, indent=2))
    print("Toxicity: " + str(json_text['attributeScores']['TOXICITY']['spanScores'][0]['score']['value']))
    print("Severe Toxicity: " + str(json_text['attributeScores']['SEVERE_TOXICITY']['spanScores'][0]['score']['value']))
    print("Identity Attack: " + str(json_text['attributeScores']['IDENTITY_ATTACK']['spanScores'][0]['score']['value']))
    print("Profanity: " + str(json_text['attributeScores']['PROFANITY']['spanScores'][0]['score']['value']))
    print("Sexually Explicit: " + str(json_text['attributeScores']['SEXUALLY_EXPLICIT']['spanScores'][0]['score']['value']))
    print("Inflammatory: " + str(json_text['attributeScores']['INFLAMMATORY']['spanScores'][0]['score']['value']))
    print("Threat: " + str(json_text['attributeScores']['THREAT']['spanScores'][0]['score']['value']))
```

```python
# Cleaning the data

def sanitize(tweet):
    tweet = str(tweet)
    tweet = re.sub("@[A-Za-z0-9]+","",tweet) #Remove @ sign
    tweet = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", tweet) #Remove http links
    tweet = " ".join(tweet.split())
    # tweet = ''.join(c for c in tweet if c not in emoji.UNICODE_EMOJI) #Remove Emojis
    tweet = tweet.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    tweet = " ".join(w for w in nltk.wordpunct_tokenize(tweet) \
         if w.lower() in words or not w.isalpha())          
    return tweet
```

```python
import requests
from ratelimit import limits, sleep_and_retry

@sleep_and_retry
@limits(calls=60, period=65)
def score_text(row):
    toxicity = 0
    severe_toxicity = 0
    identity_attack = 0
    profanity = 0
    sexually_explicit = 0
    inflammatory = 0
    threat = 0
    try:
        analyze_request = {
        'comment': { 'text': row.text},
        'requestedAttributes': {'TOXICITY': {}, 
                        'SEVERE_TOXICITY': {},
                        'IDENTITY_ATTACK': {}, 
                        'PROFANITY': {}, 
                        'SEXUALLY_EXPLICIT': {}, 
                        'INFLAMMATORY': {}, 
                        'THREAT': {} }
        }
        response = client.comments().analyze(body=analyze_request).execute()
        json_text = json.loads(json.dumps(response, indent=2))
        toxicity = json_text['attributeScores']['TOXICITY']['spanScores'][0]['score']['value']
        severe_toxicity = json_text['attributeScores']['SEVERE_TOXICITY']['spanScores'][0]['score']['value']
        identity_attack = json_text['attributeScores']['IDENTITY_ATTACK']['spanScores'][0]['score']['value']
        profanity = json_text['attributeScores']['PROFANITY']['spanScores'][0]['score']['value']
        sexually_explicit = json_text['attributeScores']['SEXUALLY_EXPLICIT']['spanScores'][0]['score']['value']
        inflammatory = json_text['attributeScores']['INFLAMMATORY']['spanScores'][0]['score']['value']
        threat = json_text['attributeScores']['THREAT']['spanScores'][0]['score']['value']
    except:
        print("Ignore")
    
    row_score = [row.text, row.Category, toxicity, severe_toxicity, identity_attack, profanity, sexually_explicit, inflammatory, threat]
    return row_score
```

For this paper, we use Perspective API developed by Jigsaw and Google's Counter Abuse Technology team to identify the toxic comments in the Twitter dataset. According to Perspective, toxicity is defined as "a rude, disrespectful, or unreasonable comment that is likely to make you leave a discussion." The API is hosted on Google Cloud Platform which can be enabled through Google Cloud Console after filling up a short form. The API uses machine learning models to identify such contents. It also provides the score of different attributes of the content. For our research paper, we will be using seven attributes: Toxicity, Severe Toxicity, Identity Attack, Profanity, Sexually Explicit, Inflammatory, and Threat. There are various experimental attributes as well which we won't be exploring. Context is also crucial while dealing with toxicity. This is, however, a feature that will be available in the future. 
The Perspective API provides a probability score for each of the attributes enabled by the user. The value of the score for each attribute ranges from 0 to 1, 0 being least toxic and 1 being the most toxic. A threshold value is chosen (0.5 in our case) to determine whether the content is toxic or not.
